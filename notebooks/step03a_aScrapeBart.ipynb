```python
# scrape 2009 to 2023

from selenium import webdriver
import pandas as pd
import io
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time

service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service)

bartdata = pd.read_csv('step03a_bart0824.csv')
bartdata = bartdata[bartdata['Year']==2008]

for y in range(2009,2024):

    driver.get(f'https://barttorvik.com/team-tables_each.php?year={y}&top=0&conlimit=All')
    driver.maximize_window()
    element = WebDriverWait(driver, 1000).until(EC.presence_of_element_located((By.XPATH, "//table")))
    with io.StringIO(driver.page_source) as f:
        df = pd.read_html(f)[0]
        bartdata = pd.concat([bartdata,df])

bartdata.to_csv('step03a_bart0824.csv',index=False)
```

```python
# add or update 2024

from selenium import webdriver
import pandas as pd
import io
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time

service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service)

y = 2024

driver.get(f'https://barttorvik.com/team-tables_each.php?year={y}&top=0&conlimit=All')
driver.maximize_window()
element = WebDriverWait(driver, 1000).until(EC.presence_of_element_located((By.XPATH, "//table")))
with io.StringIO(driver.page_source) as f:
    df = pd.read_html(f)[0]

bartdata = pd.read_csv('step03a_bart0824.csv')
bartdata = bartdata[bartdata['Year']<y]
bartdata = pd.concat([bartdata,df])
bartdata.to_csv('step03a_bart0824.csv',index=False)
```
